## Filling Null Values

```python
df['waterfront'].fillna('0.0', inplace=True)
df['view'].fillna('0.0', inplace=True)
df['yr_renovated'].fillna('0', inplace=True)
```

### Dropping ID Column

```python
cols = ['id']
df_clean = df.drop(cols, axis = 1)
df_clean.head()
```

#### Random
* '?' in df_clean[sqft_basement]
* 

```python
df_clean.loc[df_clean['sqft_basement'] == '?']
```

```python
df_clean['sqft_basement'] = df_clean['sqft_basement'].apply(lambda x: x.replace('?', '0'))
```

```python
df_clean.head()
```

```python
def get_binary_column(dfName, colName, qual_for_false):
    newCol = []
    for val in dfName[colName]:
        if val <= qual_for_false:
            newCol.append('0')
        else:
            newCol.append('1')
    return newCol
```

```python
df_clean['has_basement'] = get_binary_column(df_clean, 'sqft_basement', '0.0')
```

```python
df_clean.head()
```

```python
df_clean['is_cust_build'] = get_binary_column(df_clean, 'grade', 11)
df_clean['is_cust_build'].nunique()
```
