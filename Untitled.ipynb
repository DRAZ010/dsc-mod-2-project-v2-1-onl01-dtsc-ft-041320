```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns

from statsmodels.formula.api import ols

%matplotlib inline
```

```python
df = pd.read_csv('kc_cleaned_data.csv')
print(df.shape)
df.head()
```

# What variables work best for people wanting to flip houses?

### Bedrooms

```python
def get_years_since_renovation():
    newCol = []
    for value in df['yr_renovated']:
        num = value
        now = 2020
        since = now - num
        if since != 2020.0:
            newCol.append(since)
        else:
            newCol.append(0.0)
    return newCol
```

```python
df['years_since_renovated'] = get_years_since_renovation()
df.head()
df.loc[df['years_since_renovated'] > 0.0]
```

```python
df['is_cust_build'].value_counts().plot(kind="bar")
```

```python
outcome = 'price'
x_cols = ['years_since_renovated']
predictors = '+'.join(x_cols)
formula = outcome + '~' + predictors
model = ols(formula=formula, data=df).fit()
model.summary()
```
